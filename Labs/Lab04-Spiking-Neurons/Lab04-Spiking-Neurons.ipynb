# L04 Simulating Spiking Neurons

This notebook uses a basic leaky integrate-and-fire model to show how input current can drive a neuron toward a spike threshold.


## Leaky Integrate-and-Fire Model


```python
import numpy as np
import matplotlib.pyplot as plt

time = np.arange(0, 200, 1)
voltage = np.zeros_like(time, dtype=float)
spikes = []
threshold = 1.0
reset = 0.0
leak = 0.92
input_current = 0.16

for t in range(1, len(time)):
    voltage[t] = leak * voltage[t-1] + input_current
    if voltage[t] >= threshold:
        spikes.append(t)
        voltage[t] = reset

plt.figure(figsize=(10, 4))
plt.plot(time, voltage, label='Membrane potential')
plt.axhline(threshold, color='red', linestyle='--', label='Spike threshold')
plt.scatter(spikes, [threshold] * len(spikes), color='black', marker='x', label='Spikes')
plt.title('Leaky Integrate-and-Fire Neuron Simulation')
plt.xlabel('Time step')
plt.ylabel('Voltage')
plt.legend()
plt.tight_layout()
plt.show()
print('Spike times:', spikes)
```
