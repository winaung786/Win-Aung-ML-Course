# L05 Bio-Inspired Perception Pipeline

This notebook demonstrates a simple perception workflow using synthetic image data, edge-like filtering, and feature summarization.


## Synthetic Sensory Input


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

image = np.zeros((32, 32))
image[8:24, 14:18] = 1.0
image[14:18, 8:24] = 1.0
image += np.random.default_rng(4374).normal(0, 0.08, image.shape)
image = np.clip(image, 0, 1)

plt.imshow(image, cmap='gray')
plt.title('Synthetic Visual Input')
plt.axis('off')
plt.show()
```

## Feature Extraction


```python
vertical_edges = np.abs(np.diff(image, axis=1)).mean()
horizontal_edges = np.abs(np.diff(image, axis=0)).mean()
brightness = image.mean()
features = pd.DataFrame({
    'feature': ['mean_brightness', 'vertical_edge_energy', 'horizontal_edge_energy'],
    'value': [brightness, vertical_edges, horizontal_edges]
})
features
```
