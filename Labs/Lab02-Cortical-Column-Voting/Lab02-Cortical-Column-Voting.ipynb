# L02 Cortical Column Voting Simulation

This notebook demonstrates a simplified cortical-column voting process. Each simulated column receives noisy evidence, produces a local vote, and contributes to a final group decision.


## Setup


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(4374)
classes = ['edge', 'curve', 'texture']
columns = [f'Column_{i}' for i in range(1, 13)]
true_signal = np.array([0.25, 0.55, 0.20])
noise = np.random.normal(0, 0.12, size=(len(columns), len(classes)))
evidence = np.clip(true_signal + noise, 0, None)
probabilities = evidence / evidence.sum(axis=1, keepdims=True)
df = pd.DataFrame(probabilities, columns=classes, index=columns)
df
```

## Voting Result


```python
votes = df.idxmax(axis=1)
summary = votes.value_counts().rename_axis('class').reset_index(name='votes')
print(summary)
print('Final population decision:', summary.sort_values('votes', ascending=False).iloc[0]['class'])
summary.plot(kind='bar', x='class', y='votes', legend=False, title='Cortical Column Voting Result')
plt.ylabel('Number of Column Votes')
plt.tight_layout()
plt.show()
```
