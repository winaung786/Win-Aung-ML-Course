# L07 Attention and AI Audit Practice

This notebook creates a small audit checklist for thinking about attention, memory, and interpretability in AI systems.


## Audit Checklist


```python
import pandas as pd

audit = pd.DataFrame([
    {'dimension': 'Attention', 'question': 'What inputs does the system prioritize?', 'risk': 'Important context may be ignored.'},
    {'dimension': 'Memory', 'question': 'What information is retained across steps?', 'risk': 'The system may lose continuity or overfit to recent input.'},
    {'dimension': 'Feedback', 'question': 'How does the system update after errors?', 'risk': 'Poor feedback loops can reinforce mistakes.'},
    {'dimension': 'Explainability', 'question': 'Can a user understand the reason for the output?', 'risk': 'Opaque outputs reduce trust and accountability.'},
])
audit
```

## Simple Scoring Example


```python
scores = {'Attention': 4, 'Memory': 3, 'Feedback': 3, 'Explainability': 2}
score_table = pd.DataFrame(scores.items(), columns=['dimension', 'score_out_of_5'])
score_table['interpretation'] = score_table['score_out_of_5'].apply(lambda x: 'strong' if x >= 4 else 'needs improvement')
score_table
```
